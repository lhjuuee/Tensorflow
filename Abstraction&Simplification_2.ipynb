```python
'''
x_image = tf.reshape(x_data, [-1, 28, 28, 1])

# Conv_1 - Maxpool_1

conv1 = layers.convolution2d(x_image, 32, [5,5],
                             activation_fn=tf.nn.relu,
                             biases_initializer=tf.constant_initializer(0.1),
                             weights_initializer=tf.truncated_normal_initializer(stddev=0.1))
max_pool1 = layers.max_pool2d(conv1, [2,2])

# Conv_2 - Maxpool_2

conv2 = layers.convolution2d(max_pool1, 64, [5,5],
                             activation_fn=tf.nn.relu,
                             biases_initializer=tf.constant_initializer(0.1),
                             weights_initializer=tf.truncated_normal_initializer(stddev=0.1))
max_pool2 = layers.max_pool2d(conv2, [2,2])

# Fully connected 

max_pool2_flat = tf.reshape(max_pool2, [-1, 7*7*64])
fc1 = layers.fully_connected(max_pool2_flat, 1024,
                             activation_fn=tf.nn.relu,
                             biases_initializer=tf.constant_initializer(0.1),
                             weights_initializer=tf.truncated_normal_initializer(stddev=0.1))

# Drop out and output layer

fc1_drop = layers.dropout(fc1, keep_prob=params["dropout"],
                          is_training=(mode == 'train'))

y_conv = layers.fully_connected(fc1_Drop, 10, activation_fn=None)
'''




# Function define

def model_fn(x, target, mode, params):
    y_ = tf.cast(target, tf.float32)
    x_image = tf.reshape(x_data, [-1, 28, 28, 1])
    
    # Conv_1 - Maxpool_1

    conv1 = layers.convolution2d(x_image, 32, [5,5],
                                 activation_fn=tf.nn.relu,
                                 biases_initializer=tf.constant_initializer(0.1),
                                 weights_initializer=tf.truncated_normal_initializer(stddev=0.1))
    max_pool1 = layers.max_pool2d(conv1, [2,2])

# Conv_2 - Maxpool_2

    conv2 = layers.convolution2d(max_pool1, 64, [5,5],
                                 activation_fn=tf.nn.relu,
                                 biases_initializer=tf.constant_initializer(0.1),
                                 weights_initializer=tf.truncated_normal_initializer(stddev=0.1))
    max_pool2 = layers.max_pool2d(conv2, [2,2])

# Fully connected 

    max_pool2_flat = tf.reshape(max_pool2, [-1, 7*7*64])
    fc1 = layers.fully_connected(max_pool2_flat, 1024,
                                 activation_fn=tf.nn.relu,
                                 biases_initializer=tf.constant_initializer(0.1),
                                 weights_initializer=tf.truncated_normal_initializer(stddev=0.1))

# Drop out and output layer

    fc1_drop = layers.dropout(fc1, keep_prob=params["dropout"],
                              is_training=(mode == 'train'))

    y_conv = layers.fully_connected(fc1_drop, 10, activation_fn=None)
    
    
# Define loss function and optimization method
    
    cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y_conv, labels=y_))
    
    train_op = tf.contrib.layers.optimize_loss(loss=cross_entropy,
                                               global_step=tf.contrib.framework.get_global_step(),
                                               learning_rate=params["learning_rate"],
                                               optimizer="Adam")
    
    predictions = tf.argmax(y_conv, 1)
    
    return predictions, cross_entropy, train_op
    
    
    
```
