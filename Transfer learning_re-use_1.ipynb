# Transfer learning
Normally, training large neural network from scratch is not a good idea. First thing you need to do is finding another model

that train similar area that you want to predict. And then, reuse its structure. It is called transfer learning. It also allow you

to train much faster and use small amount of data. For example, there is neural model to classify their categories including 

car, animal or others. And let's suppose you want to classify detail type of car. Then, you can reuse part of its model. It would be

very helpful.



## 1. Re-use tensorflow model
##### In TensorFlow

### 1.1 Load model
As I mentioned before, we can restore model by import_meta_graph() function which returns saver class used to call parameters.

The model below is trained model example.


```python
import tensorflow as tf

tf.reset_default_graph() 
```

```python
saver = tf.train.import_meta_graph("./final_save_test.ckpt.meta")
```

### 1.2 Define object for training. Calculation & Tensor

For this, we use get_operation_by_name() and get_Ttensor_by_name method.

Tensor name is followed by :0. (:1 for second, :2 for third...)

```python
X = tf.get_default_graph().get_tensor_by_name("X:0")
Y = tf.get_default_graph().get_tensor_by_name("Y:0")

accuracy = tf.get_default_graph().get_tensor_by_name("eval/accuracy:0")
training_op = tf.get_default_graph().get_operation_by_name("train/Momentum")
```

### 1.3 Finding operation list

If pre-trained model is not defined well, we have to find it to get a name of it.

You can explore graph by tensorboard, or

use **get_operations()** method.

```python
for op in tf.get_default_graph().get_operations():
    print(op.name)
```

### 1.4 How to documentate calculation

To be easily reused, calculation in model need to be documentated. One good way to do it is to make **collection**

```python
for op in (X, Y, accuracy, training_op):
    tf.add_to_collection("my_importants_ops", op)
```

When it being reused,

```python
# X, Y, accuracy, training_op = tf.get_collection("my_important_ops")
```
